## Exercice 1.7 — Vectorisation
Objectif : Comprendre pourquoi NumPy est plus rapide qu'une boucle Python. 

Instructions :

- Convertissez le tableau des salaires (en BIF) en une nouvelle unité de votre choix, sans écrire de boucle for.
- Écrivez la même opération avec une boucle for classique.
- Comparez le temps d'exécution des deux approches avec le module time. <br>
**Difficulté** : 🟡 Intermédiaire

```python
# Conversion BIF vers CAD
taux_conversion = 5500
salaires_cad = salaires / taux_conversion
print(salaires_cad)
```

```python
# Caonversion avec une boucle for
salaires_cad_for = []

for salaire in salaires:
    salaire_converti = salaire / taux_conversion
    salaires_cad_for.append(salaire_converti)

salaires_cad_for
```

```python
# Temps d'execution sans boucle
debut = time.time()
salaires_usd_vectorise = salaires / taux_conversion
fin = time.time()
temps_numpy = fin - debut
print("Temps NumPy :", temps_numpy, "secondes")
```

```python
# Temps d'execution avec boucle
debut = time.time()
salaires_cad_for = []
for salaire in salaires:
    salaire_converti = salaire / taux_conversion
    salaires_cad_for.append(salaire_converti)
fin = time.time()
temps_numpy_for = fin - debut
print("Temps NumPy :", temps_numpy_for, "secondes")
```

```python
difference = temps_numpy_for / temps_numpy
print(
    f"La boucle for est environ {difference:.5f} fois plus lente"
)
```

## Exercice 1.8 — Broadcasting
Objectif : Comprendre comment NumPy applique des opérations entre tableaux de formes différentes. 

Instructions :

- Appliquez une augmentation de salaire uniforme (par exemple +10 %) à tout le tableau des salaires en une seule opération.
- Centrez-réduisez (standardisez) un tableau numérique de votre choix en soustrayant sa moyenne puis en divisant par son écart-type.
- Expliquez ce que le broadcasting a permis d'éviter. <br>
**Difficulté** : 🟡 Intermédiaire

```python
 #Augmentation de 10 %
salaires_augmentes = salaires * 1.10
salaires_augmentes
```

```python
# Calcul de la moyenne
moyenne = salaires.mean()

# Calcul de l'écart-type
ecart_type = salaires.std()

# Standardisation
salaires_standardises = (salaires - moyenne) / ecart_type

salaires_standardises
```

Le broadcasting a permis d'éviter :

- l'écriture de boucles for explicites ;
- la création de tableaux intermédiaires contenant la même valeur répétée ;
- un code plus long et moins lisible ;
- une perte de performance liée aux boucles Python.

Objectif : Relier NumPy à des concepts probabilistes de base. 

Instructions :

- Calculez, uniquement avec NumPy (booléens + mean()), la proportion d'employés satisfaits (Satisfait == "Oui") : cette proportion est une estimation d'une probabilité.
- Standardisez (z-score) la colonne Age et expliquez pourquoi cette transformation est utile avant d'entraîner un modèle.
- Discutez en une ou deux phrases du lien entre fréquence observée et probabilité en Machine Learning.<br> 
**Difficulté** : 🔴 Avancé

```python
# Conversion de la colonne Satisfait en tableau NumPy
satisfaits = employes["Satisfait"].values

# Calcul de la proportion de Oui
proportion_satisfaits = np.mean(satisfaits == "Oui")

print("Proportion d'employés satisfaits :", proportion_satisfaits)
```

```python
# Extraction de Age en tableau NumPy
ages = employes["Age"].values

# Calcul de la moyenne et de l'écart-type
moyenne_age = np.mean(ages)
ecart_type_age = np.std(ages)

# Standardisation
age_standardise = (
    ages - moyenne_age
) / ecart_type_age

print(age_standardise[:10])
print("Moyenne: ", np.mean(age_standardise))     
print("Ecart-type: ", np.std(age_standardise))
```

La standardisation est utile car les variables peuvent avoir des échelles différentes.

Par exemple :

Age : entre 20 et 60 ;
Salaire_Mensuel_BIF : centaines de milliers de BIF ;
Nombre_Absences : quelques unités.

Sans standardisation, certaines variables avec de grandes valeurs peuvent avoir une influence disproportionnée sur le modèle, notamment pour :

la régression linéaire ;
la régression logistique ;
les réseaux de neurones ;
les algorithmes utilisant les distances (KNN, clustering).

La transformation en z-score place les variables sur une échelle comparable.
