## Stratégie de traitement des valeurs manquantes

Le choix de la méthode de traitement dépend de la nature de la variable (numérique ou catégorielle) ainsi que de la distribution des données.

### 1. Colonnes numériques

Les variables numériques représentent des quantités mesurables. Plusieurs stratégies peuvent être utilisées selon les caractéristiques des données.

#### Imputation par la moyenne

La moyenne est adaptée lorsque les données suivent une distribution symétrique et qu'elles ne contiennent pas de valeurs aberrantes importantes.

**Avantages :**
- Facile à mettre en œuvre.
- Conserve la taille du jeu de données.

#### Imputation par la médiane

La médiane est préférable lorsque la variable contient des valeurs aberrantes (outliers).

**Pourquoi ?**
Contrairement à la moyenne, la médiane est peu influencée par les valeurs extrêmes. Elle fournit une estimation plus robuste.

### 2. Colonnes catégorielles

Les variables catégorielles représentent des catégories ou des modalités .

#### Imputation par la mode

La stratégie la plus courante consiste à remplacer les valeurs manquantes par la **mode**, c'est-à-dire la catégorie la plus fréquente.

**Pourquoi ?**
- Elle conserve la catégorie la plus représentative du jeu de données.
- Elle évite la création d'une nouvelle catégorie artificielle.
- Elle est simple et efficace lorsque le pourcentage de valeurs manquantes est faible.

---

## Cas de suppression des données

Lorsque le nombre de valeurs manquantes est très élevé (par exemple supérieur à 40 % ou 50 %), il peut être préférable de supprimer la colonne si elle apporte peu d'informations. De même, les lignes contenant un grand nombre de valeurs manquantes peuvent être supprimées si leur conservation risque de dégrader la qualité des analyses.

## Exercice 2.13 — Feature engineering (partie 2) — Encodage
Objectif : Préparer les variables catégorielles pour le Machine Learning (Partie 5). 

Instructions :

- Encodez la variable cible Satisfait en 0/1.
- Encodez les variables catégorielles ordinales (Niveau_Etudes, Equilibre_Vie_Travail, Satisfaction_Salaire) en respectant leur ordre logique (par exemple à l'aide d'un dictionnaire de correspondance appliqué avec .map()).
- Justifiez pourquoi ces colonnes ne doivent pas être encodées avec un simple one-hot encoding.<br> **Difficulté** : 🔴 Avancé

```python
# Encodage de la variable cible Satisfait
data_employee["Satisfait"] = data_employee["Satisfait"].map({
    "Oui": 1,
    "Non": 0
})
```

### Encodage des variables ordinales

```python
# Niveau_Etudes

niveau_etudes = {
    "Primaire": 1,
    "Secondaire": 2,
    "Licence": 3,
    "Master": 4,
    "Doctorat": 5
}

data_employee["Niveau_Etudes"] = data_employee["Niveau_Etudes"].map(niveau_etudes)
```

```python
# Equilibre_Vie_Travail

equilibre = {
    "Très faible": 1,
    "Faible": 2,
    "Moyen": 3,
    "Bon": 4,
    "Excellent": 5
}

data_employee["Equilibre_Vie_Travail"] = data_employee["Equilibre_Vie_Travail"].map(equilibre)
```

```python
# Satisfaction_Salaire

satisfaction_salaire = {
    "Très insatisfait": 1,
    "Insatisfait": 2,
    "Neutre": 3,
    "Satisfait": 4,
    "Très satisfait": 5
}

data_employee["Satisfaction_Salaire"] = data_employee["Satisfaction_Salaire"].map(satisfaction_salaire)
```

#### Justification

Les variables **Niveau_Etudes**, **Equilibre_Vie_Travail** et **Satisfaction_Salaire** sont des **variables ordinales**, c'est-à-dire que leurs catégories suivent un ordre logique.

Pour cette raison, un **encodage ordinal** est plus approprié qu'un **one-hot encoding**.

Le **one-hot encoding** transforme chaque catégorie en une colonne indépendante contenant des valeurs 0 ou 1. Cette méthode est adaptée aux variables **nominales**, mais elle supprime complètement l'information d'ordre existant entre les catégories.

En utilisant un encodage ordinal avec `.map()`, on conserve la hiérarchie naturelle des modalités, ce qui permet à de nombreux algorithmes de machine learning (régression, arbres de décision, gradient boosting, etc.) d'exploiter cette information pour améliorer leurs performances.
