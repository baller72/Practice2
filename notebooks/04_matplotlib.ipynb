### Exercice 4.11 — Boxplot : détection de valeurs aberrantes
**Objectif** : Identifier des outliers potentiels.<br> **Instructions** : Tracez un boxplot du loyer dans rent_prediction.csv. Combien de valeurs semblent aberrantes (en dehors des moustaches) ?<br> **Difficulté** : 🟡 Intermédiaire

```python
# 2. Nettoyer : supprimer les valeurs manquantes pour l'analyse (important !)
loyers = data_rent['LoyerMensuel_BIF'].dropna()

# 3. Tracer le boxplot
plt.figure(figsize=(8, 6))
plt.boxplot(loyers, vert=True, patch_artist=True)
plt.title("Boxplot des loyers")
plt.ylabel("Montant du loyer")
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.savefig("../images/boxplot_outliers.png", dpi=300, bbox_inches="tight")
plt.show()

# 4. Compter mathématiquement les outliers (méthode de l'IQR)
Q1 = loyers.quantile(0.25)
Q3 = loyers.quantile(0.75)
IQR = Q3 - Q1

borne_inferieure = Q1 - 1.5 * IQR
borne_superieure = Q3 + 1.5 * IQR

outliers = data_rent[(data_rent['LoyerMensuel_BIF'] < borne_inferieure) | (data_rent['LoyerMensuel_BIF'] > borne_superieure)]
nombre_outliers = outliers.shape[0]

print(f"Nombre total de lignes dans le jeu de données : {len(data_rent)}")
print(f"Nombre de valeurs aberrantes (outliers) détectées : {nombre_outliers}")
```

### Exercice 4.13 — Personnalisation avancée
**Objectif** : Produire un graphique de qualité "portfolio".<br> **Instructions** : Reprenez un graphique précédent et personnalisez-le entièrement : palette de couleurs cohérente, taille de figure adaptée, grille discrète, annotations sur les valeurs clés, style global (plt.style.use).<br> **Difficulté** : 🔴 Avancé

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# ------------------------------------------------------------
# 1. DONNÉES (exemple - adaptez si votre colonne s'appelle autrement)
# ------------------------------------------------------------
# salaire_moyen_par_ancienete = data_employee.groupby("Anciennete_Entreprise")["Salaire_Mensuel_BIF"].mean()

# Pour que le code soit exécutable directement, je crée un jeu de test :
# (Commentez ces 3 lignes si vous avez déjà votre Series)
np.random.seed(42)
x = np.arange(0, 11)
y = 500000 + x * 80000 + np.random.normal(0, 30000, len(x))  # Simulation
salaire_moyen_par_ancienete = pd.Series(y, index=x)

# ------------------------------------------------------------
# 2. CONFIGURATION DU STYLE GLOBAL (la base du "portfolio")
# ------------------------------------------------------------
# 'seaborn-v0_8-whitegrid' : fond blanc, grille discrète très élégante
# Si vous ne l'avez pas, installez seaborn avec : pip install seaborn
plt.style.use('seaborn-v0_8-whitegrid')

# ------------------------------------------------------------
# 3. FIGURE + PALETTE DE COULEURS
# ------------------------------------------------------------
fig, ax = plt.subplots(figsize=(14, 7))  # Taille panoramique professionnelle

# Palette dégradée : du bleu soutenu au turquoise (cohérence visuelle)
valeurs = salaire_moyen_par_ancienete.values
annees = salaire_moyen_par_ancienete.index
couleurs = plt.cm.Blues(np.linspace(0.4, 0.9, len(valeurs)))

# ------------------------------------------------------------
# 4. TRACÉ PRINCIPAL (Barres avec bordures blanches pour aérer)
# ------------------------------------------------------------
bars = ax.bar(annees, valeurs, color=couleurs, edgecolor='white', linewidth=2)

# ------------------------------------------------------------
# 5. PERSONNALISATION AVANCÉE (grille, épines, axes)
# ------------------------------------------------------------
# Grille discrète (uniquement horizontale pour éviter la surcharge)
ax.grid(axis='y', linestyle='--', alpha=0.3)
ax.grid(axis='x', linestyle='', alpha=0)  # Désactive la grille verticale

# Supprimer les bordures supérieure et droite (trend "minimaliste")
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_color('gray')
ax.spines['bottom'].set_color('gray')

# Formatage de l'axe Y (séparateurs de milliers)
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'{int(x):,}'.replace(',', ' ')))

# Étiquettes et titres
ax.set_xlabel("Ancienneté dans l'entreprise (années)", fontsize=13, labelpad=12)
ax.set_ylabel("Salaire mensuel moyen (BIF)", fontsize=13, labelpad=12)
ax.set_title("📈 Progression du salaire moyen en fonction de l'ancienneté",
             fontsize=18, fontweight='bold', pad=25)

# ------------------------------------------------------------
# 6. ANNOTATIONS DES VALEURS CLÉS (la touche "portfolio")
# ------------------------------------------------------------
# Valeurs maximales et minimales
max_val = max(valeurs)
min_val = min(valeurs)
idx_max = annees[valeurs.argmax()]
idx_min = annees[valeurs.argmin()]

# Afficher le montant AU-DESSUS de chaque barre (taille discrète)
for i, (annee, salaire) in enumerate(zip(annees, valeurs)):
    # On n'affiche pas le texte si la barre est trop petite (moins de 3% de l'écart)
    label = f"{salaire:,.0f}".replace(',', ' ')
    ax.text(annee, salaire + (max_val * 0.008), label,
            ha='center', va='bottom', fontsize=9, color='#333333', weight='semibold')

# ANNOTATION SPÉCIALE : Flèche sur le salaire MAXIMUM
ax.annotate(f'🔺 Maximum : {max_val:,.0f} BIF'.replace(',', ' '),
            xy=(idx_max, max_val),
            xytext=(idx_max + 1.2, max_val + (max_val * 0.12)),
            arrowprops=dict(arrowstyle='->', color='#D62728', lw=1.8),
            color='#D62728', fontweight='bold', fontsize=11,
            bbox=dict(boxstyle="round,pad=0.3", facecolor='white', edgecolor='#D62728', alpha=0.8))

# ANNOTATION SPÉCIALE : Encadré sur le salaire MINIMUM (pour contraste)
ax.annotate(f'⬇️ Début de carrière',
            xy=(idx_min, min_val),
            xytext=(idx_min - 1.5, min_val - (max_val * 0.1)),
            arrowprops=dict(arrowstyle='->', color='#1F77B4', lw=1.5),
            color='#1F77B4', fontweight='bold', fontsize=11)

# ------------------------------------------------------------
# 7. AJUSTEMENT FINAL (éviter les chevauchements)
# ------------------------------------------------------------
plt.tight_layout()
plt.show()
```

### Exercice 4.14 — Heatmap de corrélation
**Objectif** : Visualiser une matrice de corrélation.<br> **Instructions** : À partir de la matrice de corrélation calculée en Partie 2 (Exercice 2.14), affichez-la sous forme de heatmap avec plt.imshow ou équivalent, avec une colorbar et les valeurs annotées.<br> **Difficulté** : 🔴 Avancé

```python
fig, ax = plt.subplots(figsize=(12, 10))

# --- 1. CALCULER LA MATRICE (À exécuter avant le graphique) ---
# On sélectionne uniquement les colonnes numériques pour éviter les erreurs
corr_matrix = data_rent.select_dtypes(include=['number']).corr()

# Afficher la matrice
im = ax.imshow(corr_matrix, cmap='coolwarm', vmin=-1, vmax=1)

# Configurer les étiquettes des axes
ax.set_xticks(np.arange(len(corr_matrix.columns)))
ax.set_yticks(np.arange(len(corr_matrix.index)))
ax.set_xticklabels(corr_matrix.columns, rotation=45, ha='right', fontsize=10)
ax.set_yticklabels(corr_matrix.index, fontsize=10)

# Ajouter les valeurs ANNOTÉES dans chaque cellule
for i in range(len(corr_matrix.index)):
    for j in range(len(corr_matrix.columns)):
        valeur = corr_matrix.iloc[i, j]
        # Texte en blanc si la corrélation est forte (>0.5 ou < -0.5), sinon noir
        couleur_texte = 'white' if abs(valeur) > 0.5 else 'black'
        ax.text(j, i, f'{valeur:.2f}', ha='center', va='center', 
                color=couleur_texte, fontsize=9, weight='bold')

# Ajouter la colorbar (équivalent de la légende des couleurs)
plt.colorbar(im, shrink=0.8, label='Coefficient de corrélation')

plt.title("Matrice des corrélations (avec imshow)", fontsize=16, pad=20)
plt.tight_layout()
plt.show()
```
